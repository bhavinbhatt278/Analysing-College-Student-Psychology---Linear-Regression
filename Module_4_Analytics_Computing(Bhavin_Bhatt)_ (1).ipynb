# **4) Simple Linear Regression**

### We will perform a simple linear regression with our **dependent variable** as below:

### **Goal-oriented column**  - If you have to temporarily put your plans on hold, you make sure it is your top priority to get back on track as soon as possible.

### The purpose is to predict the behaviour amongst individuals who are more goal-oriented and how do different factors(independent variables) aid or hinder an individual's motivation to achieve goal. The following **independent variables** are chosen to predict the goal-oriented behaviour of a student.

### **1.Self_Assured** - You rarely feel insecure.

### **2.Calm_under_pressure** - You can stay calm under a lot of pressure.

### **3. Determined** - You are dedicated and focused on your goals, only rarely getting sidetracked.

### **4.Pre_planner** - When starting to work on a project, you prefer to make as many decisions upfront as possible.

### **5.Philosophical_thinker**  - You often contemplate the reasons for human existence or the meaning of life

### **6.Self_doubting** - If you make a mistake, you tend to start doubting yourself, your abilities, or your knowledge

### **Model 1 - Self_Assured versus Goal-oriented**

```r
Self_assured_Model <- lm(Goal_oriented ~ Self_assured, df_final)
summary(Self_assured_Model)

library(ggplot2)
#Socially_confident
plot1 <- ggplot(data = Self_assured_Model, aes(x = Self_assured , y = Goal_oriented)) +
  geom_point(color = "darkblue", shape = 19) +
  geom_abline(intercept = Self_assured_Model$coefficients[1], slope = Self_assured_Model$coefficients[2], color = "red", linetype = "solid") +
  labs(x = "Self_assured_Model", y = "Goal_oriented") +
  ggtitle("Relationship Between Self_assured_Model and Goal_oriented") +
  theme_minimal() +
  theme(axis.text = element_text(size = 12), axis.title = element_text(size = 14),

        panel.background = element_rect(fill = "grey"))


print(plot1)
```

### **Summary** : The Self_Assured_model above shows a positive co-relation between our dependent variable Goal_oriented and independent variable Self_Assured.R-squared value: 0.08982 and p-value of 0.0.001197 simply explains that there is a significant l variance between the two variables. Thus there is not a  strong evidence that Self_assured significantly impacts Goal_oriented in this analysis and thus a Self Assured person is more likely to be a Goal_oriented person.

## **6) Multiple Linear Regression**

###From the results obtained from all the simple linear regression models above, we will now proceed with designing multiple linear regression models using combinations of variables used in the simple linear models.

###1) The first multiple linear regression model known as **Positive_model** we created contains the four different categories of variables mainly:
###a). Self_Assured
###b) Calm_under_pressure
###c) Determined
###d) Pre_planner

```r
# Load the ggplot2 library

library(ggplot2)

# Creating a  multiple linear regression model
Positive_model <- lm(Goal_oriented  ~ Self_assured + Calm_under_pressure + Determined + Pre_planner , data = df_final)
summary(Positive_model)
# Create a scatter plot with regression line
# Create a scatter plot with regression lines
plot <- ggplot(data = df_final, aes(x = Self_assured + Calm_under_pressure + Determined + Pre_planner, y = Goal_oriented )) +
  geom_point(color = "darkblue", shape = 19) +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(x = "Goal_oriented", y = "Self_assured + Calm_under_pressure + Determined + Pre_planner") +
  ggtitle("Multiple Linear Regression: Goal_oriented vs Self_assured + Calm_under_pressure + Determined + Pre_planner") +
  theme_minimal() +
  theme(axis.text = element_text(size = 12),
        axis.title = element_text(size = 14),
        plot.title = element_text(size = 16))

print(plot)
```

### **Summary**:  It can be seen from the Positive model above that the relationship between Self_assured + Calm_under_pressure + Determined + Pre_planner variables vs Goal-oriented model is positive.
###The R square value of 0.2936 shows a 29.36% variance with the Goal-oriented It can be summarized from above that the qualitites such as Self_assured, Calm_under_pressure, Determined and Pre_Planner are positive traits for a person who is goal_oriented and increasing the chances of achieving their goals by significant amount.

### **Comparison: Positive_Model vs Determined_Model**
### The models Determined_model and Positive_model show statistical significance with very low p-values, indicating that they are valuable in explaining the variance in Goal_oriented. However, when comparing these two models, Positive_model which is the multiple linear regression model, is likely the better choice. It incorporates additional independent variables (Self_assured, Calm_under_pressure, and Pre_planner), leading to a higher R-squared value of 0.2936 vs 0.2151. This suggests that Psotive_model accounts for more variability in the dependent variable, making it a more comprehensive model for predicting Goal_oriented behavior.

### 1) The Second multiple linear regression model known as **Negative_model** we created contains the 2 different categories of variables mainly:
### a). Philosophical_thinker
### b) Self_doubting

```r
# Load the ggplot2 library

library(ggplot2)

# Creating a  multiple linear regression model
Negative_model <- lm(Goal_oriented ~ Philosophical_thinker + Self_doubting  , data = df_final)
summary(Negative_model)
# Create a scatter plot with regression line
# Create a scatter plot with regression lines
plot <- ggplot(data = df_final, aes(x = Philosophical_thinker + Self_doubting , y = Goal_oriented)) +
  geom_point(color = "darkblue", shape = 19) +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(x = "Philosophical_thinker + Self_doubting", y = "Goal_oriented") +
  ggtitle("Multiple Linear Regression: Goal_oriented vs Philosophical_thinker + Self_doubting") +
  theme_minimal() +
  theme(axis.text = element_text(size = 12),
        axis.title = element_text(size = 14),
        plot.title = element_text(size = 16))

print(plot)
```

### The low R-squared value (0.04952) suggests that this model explains only a small portion of the variability in Goal_oriented behavior. The p value of 0.05968 suggestes that there isn't much significant impact on Goal_oriented behaviour.

### **Comparison**: **Negative_model vs Philosopher_thinker_model**
###Negative_model has a slightly lower p-value (0.03541) compared to Philosopher_thinker_model (0.05968). A lower p-value suggests a somewhat stronger statistical significance.  However, it's important to note that the relationships in both models are relatively weak due to low R-squared values which explains that there might be other factors that might affect the Goal_oriented behaviour in an individual.

### **Best Multiple Model:** The best Multiple Linear Regression model to be picked is Positive_Model due to its high variance of 29.36% on the Goal_oriented behaviour. Thus the traits such as Self_assured, Calm_under_pressure, Determined, pre-planner have a positive significant effect on the Goal_oriented behaviour whereas traits such as Self_doubting and Philosopher_thinker have a negative impact although not significant effect on Goal_oriented behaviour
